```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt('loaded_500.txt')
```

```python
unique, counts = np.unique(data, return_counts=True)
```

```python
print(unique,counts)
```

```python
plt.figure(figsize=(15,8))
plt.hist(data, bins=20);
```

```python
num_sixes = (data==6).sum()
num_total= data.size
```

```python
from scipy.stats import binom
```

```python
n = np.arange(num_total)
prob_n = binom.pmf(n,num_total,1/6)
plt.figure(figsize=(15,8))
plt.plot(n,prob_n, label='Prob num')
plt.axvline(num_total/6, ls='--', lw=1, label='Mean num')
plt.axvline(num_sixes, ls=':', color='#ff7272', label='Obs num')
plt.xlabel(f'Num of sixes rolled out of total rolls {num_total}')
plt.ylabel('Probability')
plt.legend();
```

```python
d = binom(num_total, 1/6)
plt.figure(figsize=(15,8))
plt.plot(n, d.sf(n))
plt.axvline(num_sixes, ls='--')
sf = d.sf(num_sixes)
plt.axhline(sf, ls='--')
plt.xlabel('Num sixes')
plt.ylabel('SF')
print(f'Only {sf*100:.1f}% of time you would roll out this many or more sixes')
```
